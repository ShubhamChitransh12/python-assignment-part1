# Student Grade Tracker

## Task 1- Data Parsing & Profile Cleaning

## Raw Data

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

## Create Storage for Clean Data

```python
cleaned_students = []
```

## Clean and Display Data

```python
for student in raw_students:

    name = student["name"].strip().title()
    roll = int(student["roll"])
    
    marks_list = student["marks_str"].split(",")
    marks = [int(m) for m in marks_list]
    
    is_valid = all(word.isalpha() for word in name.split())
    
    if is_valid:
        print("✓ Valid name")
    else:
        print("✗ Invalid name")
    
    print("====================")
    print(f"Student : {name}")
    print(f"Roll.No : {roll}")
    print(f"Marks   : {marks}")
    print("====================")
    
    cleaned_students.append({
        "name" : name,
        "roll" : roll,
        "marks" : marks
    })
```

## Check Stored Clean Data

```python
cleaned_students
```

## Find Roll Number 103

```python
for student in cleaned_students:
    if student["roll"] == 103:
        print(student["name"].upper())
        print(student["name"].lower())
```

## Task 2- Marks Analysis Using Loops & Conditionals

## Define Data

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

## Print Subject + Marks + Grade (FOR LOOP)

```python
for i in range (len(subjects)):
    
    subject = subjects[i]
    score = marks[i]
    
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject} : {score} - Grade {grade}")
```

## Total and Average

```python
total_marks = sum(marks)
average_marks = round(total_marks/len(marks),2)

print("Total Marks :", total_marks)
print("Average Marks :", average_marks)
```

## Highest & Lowest Subject

```python
max_marks = max(marks)
min_marks = min(marks)

max_index = marks.index(max_marks)
min_index = marks.index(min_marks)

print("Highest Scorning Subject :", subjects[max_index], "-", max_marks)
print("Lowest Scorning Subject :", subjects[min_index], "-", min_marks)
```

## WHILE LOOP (Add New Subjects)

```python
new_subjects_added = 0

while True:

    subject = input("Enter subject name (or type 'done' to stop):")
   
    if subject.lower() == "done":
        break

    marks_input = input("Enter marks (0-100):")
    
    # Validate Marks
    if not marks_input.isdigit():
        print("Invalid input! Enter numeric value.")
        continue

    score = int(marks_input)
    
    if score < 0 or score > 100:
        print("Marks should be between 0 and 100.")
        continue

    # Add to lists
    subjects.append(subject)
    marks.append(score)

    new_subjects_added +=1
    print("Subject added successfully!")
```

## Final Summary After WHILE LOOP

```python
total_marks = sum(marks)
average_marks = round(total_marks / len(marks),2)

print("New Subjects Added :", new_subjects_added)
print("Updated Average Marks :", average_marks)
```

## Task 3- Class Performance Summary

## Define Class Data

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

## Print Table Header & Loop + Calculate + Print Report

```python
print("Name                |Average|Status")
print("-----------------------------------------")

pass_count = 0
fail_count = 0

topper_name = ""
topper_avg = 0

total_class_avg = 0

for student in class_data:
    name = student[0]
    marks = student[1]
    
    avg = round(sum(marks) / len(marks), 2)

   # Pass/Fail
    if avg >= 60:
        status = "Pass"
        pass_count +=1
    else:
        status = "Fail"
        fail_count +=1

   # Track Topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

   # Add to total for class average
    total_class_avg +=avg

   # Print row
    print(f"{name:<18} | {avg:>6} | {status}")
```

## Final Summary

```python
class_average = round(total_class_avg / len(class_data),2)

print("\nSummary:")
print("Passed Students :", pass_count)
print("Failed Students :", fail_count)
print("Class Topper :", topper_name, "-", topper_avg)
print("Class Average :", class_average)
```

## Task 4- String Manipulation Utility

## Define Essay

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

## Strip Whitespace

```python
clean_essay = essay.strip()

print("Clean Essay:")
print(clean_essay)
```

## Convert to Title Case

```python
title_case = clean_essay.title()

print("Title Case Essay:")
print(title_case)
```

## Count "python"

```python
python_count = clean_essay.count("python")

print("Count of python':", python_count)
```

## Replace "python" with "Python 🐍"

```python
replaced_essay = clean_essay.replace("python", "Python 🐍")

print("Replaced Essay:")
print(replaced_essay)
```

## Split into Sentences

```python
sentences = clean_essay.split(". ")

print("Sentences List:")
print(sentences)
```

## Print Numbered Sentences

```python
print("Numbered Sentences:")

for i in range (len(sentences)):
    
    sentence = sentences[i]

    # Ensure it ends with '.'
    if not sentence.endswith("."):
        sentence += "."

    print (f"{i+1}. {sentence}")
```
